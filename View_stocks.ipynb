```python
#//*************************************************************************
#//*** Visualizes the Stocks Closing Prices. It's an easy check to ensure 
#//*** the data is good.
#//*************************************************************************
import json
import requests
import time
import pandas as pd
import platform
import csv
import matplotlib.pyplot as plt

#//*********************************************************************************
#//*** Read the API keys from a JSON encoded file
#//*** Located in the ignore_folder sub directory
#//*** This Folder is added to the .gitignore file and does not show up on Github
#//*** This is Authentication Best Practices for Github
#//*********************************************************************************


#//*** Load the Stock Tickers
f = open(".\\data\\stock_tickers.json", "r")
symbols = json.loads(f.read())['symbols']
interval = 'daily'
#//*** Just get amc for testing
#symbols = [ "bb" ]
f.close()
```

```python
for symbol in symbols:
    temp_df = pd.read_csv(f"./stocks/{symbol}_{interval}.csv.zip")
    
    temp_df['date'] = pd.to_datetime(temp_df['date'])
    
    fig,ax = plt.subplots()

    ax.plot(temp_df['date'],temp_df['close'] )
    
    plt.title(f"{symbol} Closing Price")


        #plt.legend(loc='upper right',bbox_to_anchor=(1.35, 1.2))
    plt.show()
```
